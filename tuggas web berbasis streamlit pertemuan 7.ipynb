```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/content/kelayakan-pendidikan-indonesia (1).csv")

# Bersihkan data (hapus kolom kosong)
df = df.drop(columns=["Unnamed: 14"], errors="ignore")

st.title("Dashboard Kelayakan Pendidikan di Indonesia")

# Pilih Provinsi
provinsi = st.selectbox("Pilih Provinsi", df["Provinsi"].unique())
filtered = df[df["Provinsi"] == provinsi]

st.subheader(f"Data Pendidikan untuk {provinsi}")
st.write(filtered.T)

# Visualisasi Jumlah Siswa, Mengulang, Putus Sekolah
st.subheader("Visualisasi Jumlah Siswa, Mengulang, dan Putus Sekolah")
cols = ['Siswa', 'Mengulang', 'Putus Sekolah']
values = [int(filtered[col].values[0]) for col in cols]

fig, ax = plt.subplots()
ax.bar(cols, values, color=["skyblue", "orange", "red"])
ax.set_ylabel("Jumlah")
st.pyplot(fig)

# Visualisasi Kondisi Ruang Kelas
st.subheader("Kondisi Ruang Kelas")
kelas_cols = ['Ruang kelas(baik)', 'Ruang kelas(rusak ringan)', 'Ruang kelas(rusak sedang)', 'Ruang kelas(rusak berat)']
kelas_vals = [int(filtered[col].values[0]) for col in kelas_cols]

fig2, ax2 = plt.subplots()
ax2.pie(kelas_vals, labels=kelas_cols, autopct='%1.1f%%', startangle=140)
ax2.axis('equal')
st.pyplot(fig2)
```
